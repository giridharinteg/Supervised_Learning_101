```
y = data['Survived']
X = data.drop(['Survived'],axis = 1)
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_val, y_train, y_val = train_test_split(X,y, stratify = y, test_size = 0.25, random_state = 101)
```

```
from sklearn.tree import DecisionTreeClassifier
# for regression problem use decision tree regressor which will make it as a predictyion problem for continous variables
from sklearn.tree import DecisionTreeRegressor
```

```
# Model Initializor
dt_model = DecisionTreeClassifier(random_state=10)
```

```
# Model fitting
dt_model.fit(X_train, y_train)
```

```
dt_model.score(X_train,y_train)
```

```
dt_model.score(X_val,y_val)
```

```
dt_model.predict(X_val)
```

```
# change of thresghold
dt_model.predict_proba(X_val)
```

```
y_pred = dt_model.predict_proba(X_val)[:,1]
```

```
new_y = []
for i in range(len(y_pred)):
  if y_pred[i] < 0.6:
    new_y.append(0)
  else:
    new_y.append(1)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_val,new_y)
```

# Now we will do some optmizations
## first max depth

```
train_accuracy = []
validation_accuracy = []
for depth in range(1,10):
    dt_model = DecisionTreeClassifier(max_depth=depth, random_state=10)
    dt_model.fit(X_train, y_train)
    train_accuracy.append(dt_model.score(X_train, y_train))
    validation_accuracy.append(dt_model.score(X_val, y_val))
```

```
frame = pd.DataFrame({'max_depth':range(1,10), 'train_acc':train_accuracy, 'valid_acc':validation_accuracy})
frame.head()
```

```
plt.figure(figsize=(12,6))
plt.plot(frame['max_depth'], frame['train_acc'], marker='o')
plt.plot(frame['max_depth'], frame['valid_acc'], marker='o')
plt.xlabel('Depth of tree')
plt.ylabel('performance')
plt.legend()
```

# Best Depth is 8

```
train_accuracy = []
validation_accuracy = []
for leaf_nodes in range(10,50):
    dt_model = DecisionTreeClassifier(max_depth=8, max_leaf_nodes=leaf_nodes, random_state=10)
    dt_model.fit(X_train, y_train)
    train_accuracy.append(dt_model.score(X_train, y_train))
    validation_accuracy.append(dt_model.score(X_val, y_val))

frame = pd.DataFrame({'max_leaf_nodes':range(10,50), 'train_acc':train_accuracy, 'valid_acc':validation_accuracy})
frame.head()

plt.figure(figsize=(12,6))
plt.plot(frame['max_leaf_nodes'], frame['train_acc'], marker='o')
plt.plot(frame['max_leaf_nodes'], frame['valid_acc'], marker='o')
plt.xlabel('leaf nodes')
plt.ylabel('performance')
plt.legend()
```

```
parameter = 'min_samples_split'
train_accuracy = []
validation_accuracy = []
for parameter_value in range(2,10):
    dt_model = DecisionTreeClassifier(max_depth=8, max_leaf_nodes=40, min_samples_split=parameter_value, random_state=10)
    dt_model.fit(X_train, y_train)
    train_accuracy.append(dt_model.score(X_train, y_train))
    validation_accuracy.append(dt_model.score(X_val, y_val))

frame = pd.DataFrame({parameter:range(2,10), 'train_acc':train_accuracy, 'valid_acc':validation_accuracy})
frame.head()

plt.figure(figsize=(12,6))
plt.plot(frame[parameter], frame['train_acc'], marker='o')
plt.plot(frame[parameter], frame['valid_acc'], marker='o')
plt.xlabel(parameter)
plt.ylabel('performance')
plt.legend()
```

```
parameter = 'min_samples_leaf'
train_accuracy = []
validation_accuracy = []
for parameter_value in range(2,19):
    dt_model = DecisionTreeClassifier(max_depth=8, 
                                      max_leaf_nodes=25, 
                                      min_samples_split=2,
                                      min_samples_leaf = parameter_value,
                                      random_state=10)
    dt_model.fit(X_train, y_train)
    train_accuracy.append(dt_model.score(X_train, y_train))
    validation_accuracy.append(dt_model.score(X_val, y_val))

frame = pd.DataFrame({parameter:range(2,19), 'train_acc':train_accuracy, 'valid_acc':validation_accuracy})
frame.head()

plt.figure(figsize=(12,6))
plt.plot(frame[parameter], frame['train_acc'], marker='o')
plt.plot(frame[parameter], frame['valid_acc'], marker='o')
plt.xlabel(parameter)
plt.ylabel('performance')
plt.legend()
```

```
dt_model = DecisionTreeClassifier(max_depth=8, max_leaf_nodes=25, random_state=10)

#fitting the model
dt_model.fit(X_train, y_train)

#Training score
print(dt_model.score(X_train, y_train))

#Validation score
print(dt_model.score(X_val, y_val))
```
